# Question 1:
How many days does it take us per order to finish working on step 11? 

```python
# add column calculating the duration for each time a step is executed (days)
df['step_duration']= (df['date_when_order_left_status']-df['date_when_order_started_status'])/pd.Timedelta(days=1)
```

```python
df
```

```python
df.to_csv('df_with_duration.csv')
```

```python
#filter only step 11 
# get the duration of step 11 for each individual order
## since step sometimes needs to be repeated, get avg between all trials per order
```

```python
step_11_avg = df.loc[df['status_number'] == 11.0].groupby('order_number')['step_duration'].mean()
```

```python
step_11_avg=pd.DataFrame(step_11_avg)
# Avg duration of step11 per order 
step_11_avg
```

```python
step_11_avg['step_duration'].describe()
```

```python
step_11_avg.plot.hist()
```

```python
sns.set(rc={"figure.figsize":(6, 5)})
sns.histplot(step_11_avg)
```

# Question 2 : 

```python
#a. How quick the different customer agents fulfill the PreCheck (1-8)?
#b. Is there a correlation between the tickets they have and the time it takes them to complete the PreCheck?
```

```python
#a.
pre_check_duration = df.loc[df['status_category'] == "Pre-Check"].groupby(['customer_agent'])['step_duration'].mean()
pre_check_duration
```

```python
#
pre_check_duration.describe()
```

```python
# check how much time & tickets per order does every agent need
pre_check_per_order = df.loc[df['status_category'] == "Pre-Check"].groupby(['customer_agent', 'order_number']).agg({'step_duration':'sum','number_of_tickets_during_status':'count'})
pre_check_per_order                                                                       
```

```python
# how many tickets is every agent haddling during precheck (on avg)
avg_tickets_per_agent= pre_check_per_order.groupby('customer_agent')['number_of_tickets_during_status'].mean() 
avg_tickets_per_agent
```

```python
#merge both avg 
table= pd.merge(pre_check_duration, avg_tickets_per_agent, right_index = True,
               left_index = True)
```

```python
table
```

```python
table= table.rename(columns={"step_duration":"avg_days_spend", "number_of_tickets_during_status":"avg_nb_tickets_per_order"})
```
